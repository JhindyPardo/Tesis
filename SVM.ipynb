```python
from sklearn.svm import SVC
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import precision_score,precision_recall_fscore_support
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_predict

#leo variables
dataset = pd.read_csv("Features unigrama.csv", sep=",", low_memory=False)
y = dataset.Label
X = dataset.drop('Label', axis=1)
scaler = MinMaxScaler(feature_range=(0, 1))
X = scaler.fit_transform(X)
svclassifier = SVC(C=1.0, kernel='linear', degree=3, gamma='auto')
# scores = cross_val_score(svclassifier, X, y, cv=5, scoring='precision_micro')
y_pred = cross_val_predict(svclassifier, X, y, cv=8, n_jobs=4)


print(confusion_matrix(y,y_pred))


#Calculo la precisión del modelo
print(metrics.accuracy_score(y,y_pred))

print(classification_report(y,y_pred))
df5 = pd.DataFrame(classification_report(y, y_pred,output_dict=True))
df5
```

```python
from sklearn.svm import SVC
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import precision_score
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_predict

#leo variables
dataset = pd.read_csv("features bigram.csv", sep=",", low_memory=False)
y = dataset.Label
X = dataset.drop('Label', axis=1)
scaler = MinMaxScaler(feature_range=(0, 1))
X = scaler.fit_transform(X)
svclassifier = SVC(C=1.0, kernel='linear', degree=3, gamma='auto')
# scores = cross_val_score(svclassifier, X, y, cv=5, scoring='precision_micro')
y_pred = cross_val_predict(svclassifier, X, y, cv=8, n_jobs=4)


print(confusion_matrix(y,y_pred))


#Calculo la precisión del modelo
print(metrics.accuracy_score(y,y_pred))

print(classification_report(y,y_pred))

df6 = pd.DataFrame(classification_report(y, y_pred,output_dict=True))
df6
```

```python
with pd.ExcelWriter("output.xlsx", engine="openpyxl", mode="a") as writer:
    df5.to_excel(writer, sheet_name='SVM unigramas')
    df6.to_excel(writer, sheet_name='SVM bigramas')
```
