```python
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_predict
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_predict
from sklearn.utils import shuffle
#Defino el algoritmo a utilizar
#Naive Bayes
algoritmo = MultinomialNB(alpha=1.0, fit_prior=True)
#leo variables
X_train = pd.read_csv("X_train_unigram.csv", sep=",")
X_test = pd.read_csv("X_test_unigram.csv", sep=",")
y_train = pd.read_csv("y_train_unigram.csv", sep=",")
y_test = pd.read_csv("y_test_unigram.csv", sep=",")
scaler = MinMaxScaler(feature_range=(0, 1))
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
#Entreno el modelo
algoritmo.fit(X_train, y_train)
#Realizo una predicción
y_pred = algoritmo.predict(X_test)
#Verifico la matriz de Confusión
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz)
#Calculo la precisión del modelo
precision = precision_score(y_test, y_pred,average='weighted')
print('Precisión del modelo:')
print(precision)

df1 = pd.DataFrame(classification_report(y_test, y_pred,output_dict=True))
df1
```

```python
#naive bayes Bigramas
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_predict
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_predict
from sklearn.utils import shuffle
#Defino el algoritmo a utilizar
#Naive Bayes
algoritmo = MultinomialNB(alpha=1.0, fit_prior=True)
#leo variables 
X_train = pd.read_csv("X_train_bigram.csv", sep=",")
X_test = pd.read_csv("X_test_bigram.csv", sep=",")
y_train = pd.read_csv("y_train_bigram.csv", sep=",").values.ravel()
y_test = pd.read_csv("y_test_bigram.csv", sep=",").values.ravel()
scaler = MinMaxScaler(feature_range=(0, 1))
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
#Entreno el modelo
algoritmo.fit(X_train, y_train)
#Realizo una predicción
y_pred = algoritmo.predict(X_test)
#Verifico la matriz de Confusión
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz)
#Calculo la precisión del modelo
precision = precision_score(y_test, y_pred,average='weighted')
print('Precisión del modelo:')
print(precision)

df2 = pd.DataFrame(classification_report(y_test, y_pred,output_dict=True))
df2
```

```python
with pd.ExcelWriter('output.xlsx') as writer:  
    df1.to_excel(writer, sheet_name='NB unigramas')
    df2.to_excel(writer, sheet_name='NB bigramas')
```
